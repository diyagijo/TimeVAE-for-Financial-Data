```python

## 1. INSTALL LIBRARIES
!pip install yfinance

## 2. IMPORTS
import yfinance as yf
import numpy as np
import pandas as pd
import tensorflow as tf
from tensorflow.keras import layers
from tensorflow.keras.models import Model
from tensorflow.keras.losses import MeanSquaredError, BinaryCrossentropy # <-- IMPORTED BCE
from tensorflow.keras.optimizers import Adam
from sklearn.preprocessing import MinMaxScaler
import matplotlib.pyplot as plt

print("All libraries imported successfully.")
print(f"TensorFlow Version: {tf.__version__}")

## 3. --- PHASE I: DATA ACQUISITION & PREPROCESSING ---

# --- Parameters ---
stock_tickers = ['AAPL', 'GOOGL', 'MSFT', 'AMZN', 'META', 'TSLA', 'NVDA', 'JPM', 'JNJ', 'V']
start_date = '2020-01-01'
end_date = '2024-12-31'
seq_len = 24  # Sequence length (e.g., 24 trading days)

# --- Data Sourcing ---
print("Fetching financial data...")
all_data = []
for ticker in stock_tickers:
    data = yf.download(ticker, start=start_date, end=end_date)
    all_data.append(data[['Open', 'High', 'Low', 'Close', 'Volume']])

# Combine and align data
raw_df = pd.concat(all_data, axis=1)
raw_df.columns = [f'{ticker}_{col}' for ticker in stock_tickers for col in ['Open', 'High', 'Low', 'Close', 'Volume']]
raw_df.dropna(inplace=True)

print("Data fetched successfully.")

# --- Normalization ---
scaler = MinMaxScaler()
scaled_data = scaler.fit_transform(raw_df)
print(f"\nData scaled. Shape: {scaled_data.shape}")

# --- Windowing (Sequencing) ---
processed_data = []
for i in range(len(scaled_data) - seq_len):
    processed_data.append(scaled_data[i:i + seq_len])

processed_data = np.array(processed_data).astype(np.float32)
print(f"Data sequenced. Final shape: {processed_data.shape}")

# --- Create TensorFlow Dataset ---
batch_size = 128
dataset = tf.data.Dataset.from_tensor_slices(processed_data)
dataset = dataset.shuffle(buffer_size=len(processed_data)).batch(batch_size).prefetch(tf.data.AUTOTUNE)

## 4. --- PHASE II: BUILD THE "MORE POWER" TIMEVAE ---

# --- Model Parameters ---
num_samples, seq_len, num_features = processed_data.shape
latent_dim = 64  # Keep wide bottleneck

# --- Build the Sampling Layer ---
class Sampling(layers.Layer):
    """Uses (z_mean, z_log_var) to sample z, the vector encoding a sequence."""
    def call(self, inputs):
        z_mean, z_log_var = inputs
        batch = tf.shape(z_mean)[0]
        dim = tf.shape(z_mean)[1]
        epsilon = tf.keras.backend.random_normal(shape=(batch, dim))
        return z_mean + tf.exp(0.5 * z_log_var) * epsilon

# --- 1. The Encoder ---
encoder_inputs = layers.Input(shape=(seq_len, num_features))
# --- Keep high capacity model ---
x = layers.GRU(256, return_sequences=True)(encoder_inputs)
x = layers.GRU(128)(x)
x = layers.Dense(64, activation="relu")(x)
z_mean = layers.Dense(latent_dim, name="z_mean")(x)
z_log_var = layers.Dense(latent_dim, name="z_log_var")(x)
z = Sampling()([z_mean, z_log_var])
encoder = Model(encoder_inputs, [z_mean, z_log_var, z], name="encoder")
encoder.summary()

# --- 2. The Decoder ---
latent_inputs = layers.Input(shape=(latent_dim,))
x = layers.RepeatVector(seq_len)(latent_inputs)
x = layers.Dense(64, activation="relu")(x)
x = layers.GRU(128, return_sequences=True)(x)
x = layers.GRU(256, return_sequences=True)(x)
decoder_outputs = layers.TimeDistributed(layers.Dense(num_features, activation="sigmoid"))(x)
decoder = Model(latent_inputs, decoder_outputs, name="decoder")
decoder.summary()

# --- 3. The Full VAE Model (with KL Annealing) ---
class VAE(Model):
    def __init__(self, encoder, decoder, **kwargs):
        super(VAE, self).__init__(**kwargs)
        self.encoder = encoder
        self.decoder = decoder
        self.total_loss_tracker = tf.keras.metrics.Mean(name="total_loss")
        self.recon_loss_tracker = tf.keras.metrics.Mean(name="recon_loss")
        self.kl_loss_tracker = tf.keras.metrics.Mean(name="kl_loss")
        self.beta = tf.Variable(0.0, trainable=False)

    @property
    def metrics(self):
        return [self.total_loss_tracker, self.recon_loss_tracker, self.kl_loss_tracker]

    def train_step(self, data):
        with tf.GradientTape() as tape:
            z_mean, z_log_var, z = self.encoder(data)
            reconstruction = self.decoder(z)

            # --- FIX #1: Use BinaryCrossentropy for a stronger loss ---
            # We multiply by num_features (50) to make its scale comparable to the KL loss
            recon_loss = BinaryCrossentropy()(data, reconstruction) * num_features

            kl_loss = -0.5 * (1 + z_log_var - tf.square(z_mean) - tf.exp(z_log_var))
            kl_loss = tf.reduce_mean(tf.reduce_sum(kl_loss, axis=1))

            total_loss = recon_loss + (self.beta * kl_loss)

        grads = tape.gradient(total_loss, self.trainable_weights)
        self.optimizer.apply_gradients(zip(grads, self.trainable_weights))

        self.total_loss_tracker.update_state(total_loss)
        self.recon_loss_tracker.update_state(recon_loss)
        self.kl_loss_tracker.update_state(kl_loss)

        return {m.name: m.result() for m in self.metrics}

print("\nTimeVAE Model built successfully.")

## 5. --- PHASE III: TRAIN THE MODEL (WITH FIX) ---

epochs = 400 # <-- FIX #2: More epochs
annealing_epochs = 150 # <-- FIX #2: More aggressive annealing

class KLAnnealingCallback(tf.keras.callbacks.Callback):
    def __init__(self, annealing_epochs):
        super(KLAnnealingCallback, self).__init__()
        self.annealing_epochs = annealing_epochs

    def on_epoch_end(self, epoch, logs=None):
        if epoch < self.annealing_epochs:
            new_beta = tf.cast(epoch / self.annealing_epochs, tf.float32)
            self.model.beta.assign(new_beta)
        else:
            self.model.beta.assign(1.0)

        if (epoch + 1) % 10 == 0:
            print(f" - Current KL Beta: {self.model.beta.numpy():.4f}")

# --- Compile and Train ---
vae = VAE(encoder, decoder)
vae.compile(optimizer=Adam(learning_rate=0.0005))

print(f"Starting training for {epochs} epochs with KL Annealing...")
history = vae.fit(dataset, epochs=epochs, callbacks=[KLAnnealingCallback(annealing_epochs)])

print("Training complete.")

# --- Plot the loss ---
plt.figure(figsize=(10, 6))
plt.plot(history.history['total_loss'], label='Total Loss')
plt.plot(history.history['recon_loss'], label='Reconstruction Loss')
plt.plot(history.history['kl_loss'], label='KL Loss')
plt.title('TimeVAE Training Loss (with KL Annealing)')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()
plt.show()

## 6. --- PHASE IV: GENERATE SYNTHETIC DATA ---
print("\nGenerating new synthetic data...")

num_synthetic_samples = len(processed_data)

random_latent_samples = tf.random.normal(
    shape=(num_synthetic_samples, latent_dim)
)
synthetic_data_scaled = decoder.predict(random_latent_samples)
synthetic_data_flattened = synthetic_data_scaled.reshape(-1, num_features)
synthetic_data = scaler.inverse_transform(synthetic_data_flattened)
synthetic_df = pd.DataFrame(synthetic_data, columns=raw_df.columns)

print("Synthetic data generated and rescaled successfully!")
print(synthetic_df.head())


## 7. --- PHASE V: EVALUATE THE SYNTHETIC DATA ---
from sklearn.manifold import TSNE

print("\nRunning t-SNE evaluation... this may take a minute.")

sample_size = 500
real_sample_flat = processed_data[:sample_size].reshape(sample_size, -1)
synth_sample_flat = synthetic_data_scaled[:sample_size].reshape(sample_size, -1)

combined_data = np.concatenate((real_sample_flat, synth_sample_flat), axis=0)
labels = np.array(['Real'] * sample_size + ['Synthetic'] * sample_size)

tsne = TSNE(n_components=2, verbose=1, perplexity=40, n_iter=300, random_state=42)
tsne_results = tsne.fit_transform(combined_data)

print("t-SNE complete. Plotting results...")

plt.figure(figsize=(12, 8))
plt.scatter(tsne_results[labels == 'Real', 0], tsne_results[labels == 'Real', 1], label='Real Data', alpha=0.6)
plt.scatter(tsne_results[labels == 'Synthetic', 0], tsne_results[labels == 'Synthetic', 1], label='Synthetic Data', alpha=0.6)
plt.title('t-SNE Visualization of Real vs. Synthetic Data (TimeVAE)')
plt.xlabel('t-SNE feature 1')
plt.ylabel('t-SNE feature 2')
plt.legend()
plt.show()

print("\n--- Project Execution Complete ---")
```

```python
## 8. --- PHASE VI: FINAL EVALUATION (DOWNSTREAM UTILITY) ---
from tensorflow.keras.layers import LSTM, Dense, Input
from tensorflow.keras.models import Model
from sklearn.metrics import mean_squared_error
import numpy as np

print("Starting Downstream Utility Test...")

# --- 1. Define the Prediction Task ---
# We will predict the next day's 50 features given the previous 23 days.

# Prepare real data
X_real = processed_data[:, :-1, :]  # All samples, first 23 steps
y_real = processed_data[:, -1, :]   # All samples, last (24th) step

# Prepare synthetic data
X_synth = synthetic_data_scaled[:, :-1, :] # All samples, first 23 steps
y_synth = synthetic_data_scaled[:, -1, :]  # All samples, last (24th) step

# --- 2. Create Training and Test Sets ---
# We'll use 10% of real data for training, and the last 20% for testing.
train_idx = int(len(X_real) * 0.10)
test_idx = int(len(X_real) * 0.80)

# Real data for training (tiny subset)
X_train_real_small = X_real[:train_idx]
y_train_real_small = y_real[:train_idx]

# Real data for testing (held-out set)
X_test_real = X_real[test_idx:]
y_test_real = y_real[test_idx:]

print(f"Baseline Training Set Size: {len(X_train_real_small)} samples")
print(f"Real Test Set Size: {len(X_test_real)} samples")

# --- 3. Build the "Downstream" Prediction Model ---
def build_predictor():
    # A simple, standard LSTM model for forecasting
    inputs = Input(shape=(seq_len - 1, num_features)) # Input shape (23, 50)
    x = LSTM(64, activation='relu')(inputs)
    x = Dense(128, activation='relu')(x)
    outputs = Dense(num_features)(x) # Output shape (50)

    model = Model(inputs, outputs)
    model.compile(optimizer=Adam(learning_rate=0.001), loss='mean_squared_error')
    return model

# --- 4. Train the BASELINE Model ---
print("\nTraining Baseline Model (Real Data Only)...")
model_baseline = build_predictor()
model_baseline.fit(
    X_train_real_small,
    y_train_real_small,
    epochs=50,
    batch_size=32,
    verbose=0, # 0 = silent training
    validation_data=(X_test_real, y_test_real)
)
print("Baseline Model trained.")

# --- 5. Train the AUGMENTED Model ---
print("Training Augmented Model (Real + Synthetic Data)...")

# Combine the small real set with all the synthetic data
X_train_augmented = np.concatenate((X_train_real_small, X_synth), axis=0)
y_train_augmented = np.concatenate((y_train_real_small, y_synth), axis=0)

# Shuffle the augmented data
np.random.seed(42) # for reproducibility
shuffle_indices = np.random.permutation(len(X_train_augmented))
X_train_augmented = X_train_augmented[shuffle_indices]
y_train_augmented = y_train_augmented[shuffle_indices]

print(f"Augmented Training Set Size: {len(X_train_augmented)} samples")

model_augmented = build_predictor()
model_augmented.fit(
    X_train_augmented,
    y_train_augmented,
    epochs=50,
    batch_size=32,
    verbose=0, # 0 = silent training
    validation_data=(X_test_real, y_test_real)
)
print("Augmented Model trained.")

# --- 6. Final Evaluation & Comparison ---
print("\n--- Final Results ---")

# Evaluate Baseline Model
preds_baseline = model_baseline.predict(X_test_real)
mse_baseline = mean_squared_error(y_test_real, preds_baseline)
print(f"Baseline Model (10% Real Data) MSE on Test Set: {mse_baseline:.6f}")

# Evaluate Augmented Model
preds_augmented = model_augmented.predict(X_test_real)
mse_augmented = mean_squared_error(y_test_real, preds_augmented)
print(f"Augmented Model (10% Real + Synthetic) MSE on Test Set: {mse_augmented:.6f}")

# --- Conclusive Result ---
if mse_augmented < (mse_baseline * 0.95):
    print("\n SUCCESS: The augmented model performed significantly better!")
    print("This proves your synthetic data adds real predictive value.")
elif mse_augmented < (mse_baseline * 1.1):
    print("\n GOOD: The augmented model performed comparably (or slightly better).")
    print("This proves your synthetic data is a high-quality, valid substitute for real data.")
else:
    print("\n❌ NOTE: The augmented model performed worse.")
    print("This suggests the synthetic data, while visually similar, may be adding noise.")
```
